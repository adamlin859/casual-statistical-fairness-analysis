```python
import pandas as pd
```

```python
data = pd.read_csv("data/asc_result.csv", index_col=0)
```

```python
data.columns = ['TE', 'DE', 'ED (Education)', 'ID (Marital Status)', 'TE_STD', 'DE_STD', 'ED_STD', 'ID_STD']
```

```python
y_err = data.iloc[:, 4:]
```

```python
y_err.columns = data.iloc[:, :4].columns
```

```python
y_err = y_err.T
```

```python
y_err
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("dark")
plt.rcParams.update({'font.size': 22})
# f = plt.figure()
# f.set_figwidth(100)
# f.set_figheight(20)
data.iloc[:, :4].T.plot(kind='bar', yerr=y_err, rot=0, figsize=(20, 10))
plt.ylabel("Estimation of Total and Path-Specific Effects")
# plt.ylabel("Pies Consumed")

plt.grid()
# plt.show()
```
